```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
df.head(12)
```

```python
x = df['month_number']
y= df['total_profit']
plt.figure(figsize=(12,5))
plt.plot(x,y)
plt.xticks(x)
plt.xlabel("month number")
plt.ylabel("total profit")
plt.title("company profit per month")
plt.show()
```

```python
x = df['month_number']
y= df['total_profit']
plt.figure(figsize=(12,5))
plt.plot(x,y,color="r",marker = 'o', ms = 12, mec = 'r',mfc = 'k', linestyle = '--',linewidth = '4.5', label = 'profit of last year')
plt.xticks(x)
plt.legend()
plt.xlabel("month number")
plt.ylabel("total profit")
plt.title("company sales data of last year")
```

```python
a = df['month_number']
b = df['facecream']
c = df['facewash']
d = df['toothpaste']
e = df['bathingsoap']
f = df['shampoo']
g = df['moisturizer']
plt.xticks(a)
plt.title("sales data")
plt.plot(a,b , label = "face cream sales data" , color = '#7FFFD4' , marker = 'o')
plt.plot(a,c , label = "face wash sales data" , color = '#556B2F' , marker = 'o')
plt.plot(a,d , label = "tooth paste sales data" , color = '#00BFFF' , marker = 'o')
plt.plot(a,e , label = "bathing soap sales data" , color = '#D2691E' , marker = 'o')
plt.plot(a,f , label = "shampoo sales data" , color = '#DC143C' , marker = 'o')
plt.plot(a,g , label = "moisturizer sales data" , color = '#DDA0DD' , marker = 'o')
plt.legend()
```

```python
x = df['month_number']
y = df['toothpaste']
plt.title("tooth paste sales data")
plt.xticks(x)
plt.scatter(x, y,label = 'tooth paste sales data')
plt.legend()
plt.grid(linestyle = '--', linewidth = 1)
plt.xlabel("month number")
plt.ylabel("number of units sold")
plt.show()
```

```python
a = df['month_number']
b = df['facecream']
c = df['facewash']
plt.title("face wash and face cream sales data")
plt.bar(a-0.2,b, width = 0.4,label = 'face cream sales data')
plt.bar(a+0.2,c, width = 0.4,label = 'face wash sales data')
plt.xlabel("month number")
plt.ylabel("sales untis in number")
ax = plt.gca()
ax.set_facecolor('#F8F8FF')
ax.grid(color='#DCDCDC')
plt.legend()
plt.xticks(a)
plt.grid(linestyle = '--', linewidth = 1)
plt.figure(figsize=(500,150))
```

```python
a = df['month_number']
b = df['bathingsoap']
plt.grid(linestyle = '--', linewidth = 1)
plt.title("bathingsoap sales data")
plt.xlabel("month number")
plt.ylabel("sales untis in number")
plt.xticks(a)
plt.bar(a,b)
```

```python

b = df['total_profit']
plt.hist(b, bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000])
plt.xlabel('Profit Range')
plt.ylabel('Actual Profit')
plt.title('Profit Data')
plt.show()
```

```python
fig = plt.figure()
b = df['facecream'].sum()
c = df['facewash'].sum()
d = df['toothpaste'].sum()
e = df['bathingsoap'].sum()
f = df['shampoo'].sum()
g = df['moisturizer'].sum()
plt.pie([b,c,d,e,f,g],labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.title('Sales Data')
plt.legend(loc='lower right')
plt.show()
langs = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo','moisturizer']
data = np.array([b, c, d, e, f, g])
ax.pie(data, labels = langs, autopct='%1.2f%%')
plt.show()
```

```python
a = df['month_number']
b = df['facecream']
c = df['facewash']
d = df['toothpaste']
e = df['bathingsoap']
f = df['shampoo']
g = df['moisturizer']
plt.stackplot(a,b,c,d,e,f,g,labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.legend(loc='upper left')
plt.title('All Product Sales Data using Stack Plot')
plt.show()
```
