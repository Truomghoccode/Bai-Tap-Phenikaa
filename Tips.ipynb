```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
tips.head()
```

```python
tips = tips.drop('Unnamed: 0', axis=1)
```

```python
plt.hist(tips['total_bill'])
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip by Size')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()
```

```python
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Tip by Day and Sex')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day and Time')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2)
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], ax=axs[0])
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], ax=axs[1])
axs[0].set_title('Dinner')
axs[1].set_title('Lunch')
plt.suptitle('Tip by Time of Day')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2)
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Male'], ax=axs[0])
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Female'], ax=axs[1])
axs[0].set_title('Male')
axs[1].set_title('Female')
plt.suptitle('Total Bill vs Tip by Sex and Smoker')
plt.show()
```
