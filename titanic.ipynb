```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv')
titanic.head()
```

```python
titanic.set_index('PassengerId', inplace=True)
```

```python
sex_counts = titanic['Sex'].value_counts()
sex_counts.plot(kind='pie', autopct='%.1f%%')
plt.title('Male/Female Proportion')
plt.show()
```

```python
colors = {'male': 'blue', 'female': 'red'}
titanic.plot(kind='scatter', x='Age', y='Fare', c=titanic['Sex'].map(colors))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age by Gender')
plt.show()
```

```python
titanic['Fare'].plot(kind='hist')
plt.xlabel('Fare')
plt.title('Histogram of Fare Paid')
plt.show()
```
