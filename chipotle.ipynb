```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv", sep='\t')
chipo = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv", sep='\t')
chipo.head(10)
```

```python
top_5 = df['item_name'].value_counts().head(5)
top_5.plot(kind='bar')
plt.xlabel('Item')
plt.ylabel('Number of Orders')
plt.title('Top 5 Items Bought')
plt.show()
```

```python
df['item_price'] = df['item_price'].str.replace('$', '').astype(float)
orders = df.groupby('order_id').sum()
plt.scatter(x=orders['item_price'], y=orders['quantity'])
plt.xlabel('Giá đơn hàng')
plt.ylabel('Số lượng mặt hàng')
plt.title('Số lượng mặt hàng so với giá đơn hàng')
plt.show()
```
